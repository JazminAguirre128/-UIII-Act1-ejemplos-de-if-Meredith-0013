## Ejemplos de for

```python
# Ejemplo 1:
numero = int(input("Ingresa un número para ver su tabla: ")) # Pide un número al usuario
for i in range(1, 11):                                      # Itera del 1 al 10
    print(f"{numero} x {i} = {numero * i}")                 # Calcula y muestra el resultado
```

```python
# Ejemplo 2:
cantidad = int(input("¿Cuántos números deseas sumar? ")) # Define el límite del ciclo
suma = 0                                                # Inicializa el acumulador
for i in range(cantidad):                               # Se repite 'cantidad' veces
    num = float(input(f"Ingresa el número {i+1}: "))     # Pide cada número
    suma += num                                         # Lo suma al total
print(f"La suma total es: {suma}")                      # Muestra el resultado final
```

```python
# Ejemplo 3:
n = int(input("¿Cuántos artículos deseas registrar? ")) # Cantidad de elementos
for i in range(n):                                      # Itera n veces
    nombre = input(f"Nombre del artículo {i+1}: ")       # Entrada de texto
    precio = float(input("Precio: "))                   # Entrada numérica
    print(f"Registrado: {nombre} - ${precio}")          # Confirmación
```

```python
# Ejemplo 4:
elementos = ["Python", "Java", "C++", "JavaScript"]         # Lista de datos
objetivo = input("¿Qué lenguaje buscas en la lista? ")      # Dato a buscar
encontrado = False                                          # Bandera de estado
for item in elementos:                                      # Recorre la lista
    if item == objetivo:                                    # Compara cada elemento
        print(f"{objetivo} está en la lista.")              # Mensaje de éxito
        encontrado = True                                   # Cambia el estado
        break                                               # Detiene el ciclo
if not encontrado:                                          # Si termina sin éxito
    print("No se encontró el lenguaje.")
```

```python
# Ejemplo 5:
frase = input("Ingresa una frase: ")              # Entrada de cadena de texto
vocales = "aeiou"                                 # Definición de criterios
contador = 0                                      # Inicializa contador
for letra in frase.lower():                       # Recorre cada letra de la frase
    if letra in vocales:                          # Verifica si es una vocal
        contador += 1                             # Incrementa si coincide
print(f"Tu frase tiene {contador} vocales.")      # Resultado
```
