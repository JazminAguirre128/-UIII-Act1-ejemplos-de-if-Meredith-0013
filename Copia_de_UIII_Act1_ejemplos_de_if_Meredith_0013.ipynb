```python
# Ejemplo 1: Sentencia if simple (Uso de Aritmética y Relacionales)
# Este código verifica si un número es par utilizando el operador de módulo y el de igualdad.
numero = 10 # Se asigna el valor 10 a la variable
if numero % 2 == 0: # Si el resto de dividir numero entre 2 es igual a 0
    print("El número es par.") # Se imprime este mensaje si la condición es verdadera
```

```python
# Ejemplo 2: Sentencia if-else (Uso de Lógicos y Relacionales)
# Aquí se determina si una persona está en edad laboral activa utilizando un rango lógico.
edad = 25 # Define la edad del usuario
if edad >= 18 and edad <= 65: # Verifica si está entre 18 y 65 años inclusive
    print("Persona en edad laboral.") # Mensaje si cumple ambas condiciones
else: # Bloque que se ejecuta si la condición de arriba es falsa
    print("Persona fuera del rango laboral estándar.") # Mensaje si es menor de 18 o mayor de 65
```

```python
# Ejemplo 2: Sentencia if-else (Uso de Lógicos y Relacionales)
# Aquí se determina si una persona está en edad laboral activa utilizando un rango lógico.
edad = 15 # Define la edad del usuario
if edad >= 18 and edad <= 65: # Verifica si está entre 18 y 65 años inclusive
    print("Persona en edad laboral.") # Mensaje si cumple ambas condiciones
else: # Bloque que se ejecuta si la condición de arriba es falsa
    print("Persona fuera del rango laboral estándar.") # Mensaje si es menor de 18 o mayor de 65
```

```python
# Ejemplo 3: Sentencia elif (Uso de Aritméticos, Lógicos y Relacionales)
# Este ejemplo es un sistema de clasificación de notas basado en múltiples rangos y cálculos.
puntaje = 85 # Variable con el puntaje obtenido
total_posible = 100 # Variable con el máximo puntaje
porcentaje = (puntaje / total_posible) * 100 # Cálculo aritmético del porcentaje

if porcentaje >= 90: # Primera condición: excelente
    print("Calificación: A")
elif porcentaje >= 80 and porcentaje < 90: # Segunda condición: muy bueno
    print("Calificación: B")
elif porcentaje >= 70 and porcentaje < 80: # Tercera condición: bueno
    print("Calificación: C")
else: # Si no cumple ninguna de las anteriores
    print("Calificación: Necesita mejorar")
```

```python
# Ejemplo 4: Identificación de Año Bisiesto (Lógica Compleja)
# Este ejemplo combina varios tipos de operadores para resolver un problema de calendario.
anio = 2024 # Año a evaluar
if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0): # Lógica de bisiesto
    print(f"{anio} es un año bisiesto.") # Mensaje si es bisiesto
else:
    print(f"{anio} no es un año bisiesto.") # Mensaje si no lo es
```
